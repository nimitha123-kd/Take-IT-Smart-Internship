2. Bank Transaction Simulator
User can:
● Deposit
● Withdraw
● Check balance
● Exit
Use:
● while True
● dictionary for accounts
● functions for operations
● conditions for insufficient balance




```python
accounts = {
     "Aman": 5000
  }

def deposit(balance, amount):
    return balance + amount

def withdraw(balance, amount):
    if balance >= amount:
        return balance - amount
    else:
        print("Insufficient Balance!")

def check_balance(name):
    print("Balance:", accounts[name])

name = input("Enter account name: ")

if name not in accounts:
    print("Account not found!")
else:
    while True:
        print("\n1.Deposit 2.Withdraw 3.Check Balance 4.Exit")
        choice = input("Enter choice: ")

        if choice == "1":
          amt = int(input("Amount: "))
          accounts["Aman"] = deposit(accounts["Aman"], amt)

        elif choice == "2":
          amt = int(input("Amount: "))
          accounts["Aman"] = withdraw(accounts["Aman"], amt)

        elif choice == "3":
          check_balance(name)

        elif choice == "4":
          print("Exiting...")
          break

        else:
            print("Invalid choice, Try Again!")
            break

```

4. Matrix Multiplication (Nested Loop Heavy)
Multiply two 3×3 matrices using nested loops only. Or  print table of 2 to 10

7. Password Strength Checker
Function checks password rules:
● ≥ 8 chars
● Uppercase
● Lowercase
● Digit
● Special symbol
Return Strong / Medium / Weak.

```python

def check_password(pwd):
    checks = {
        "length": len(pwd) >= 8,
        "upper": any(c.isupper() for c in pwd),
        "lower": any(c.islower() for c in pwd),
        "digit": any(c.isdigit() for c in pwd),
        "special": any(not c.isalnum() for c in pwd)
    }

    score = sum(checks.values())

    if score == 5:
        return "Strong"
    elif score >= 3:
        return "Medium"
    else:
        return "Weak"
# User input line
password = input("Enter your password: ")

#  Function call
result = check_password(password)
print("Password strength:", result)


#If you don’t want the password to be visible while typing:
#import getpass
#password = getpass.getpass("Enter your password: ")
```

● Allow multiple voters

```python
votes = {"Alice": 0, "Bob": 0, "Charlie": 0}
voters = set()

while True:
    voter_id = input("Enter voter ID (or exit): ")
    if voter_id == "exit":
        break

    if voter_id in voters:
        print("Duplicate vote not allowed")
        continue

    voters.add(voter_id)
    choice = input("Vote for Alice / Bob / Charlie: ")

    if choice in votes:
        votes[choice] += 1
        print("Vote recorded")
    else:
        print("Invalid candidate")

max_votes = max(votes.values())
winners = [c for c in votes if votes[c] == max_votes]

if len(winners) == 1:
    print("Winner:", winners[0])
else:
    print("Tie between:", winners)
```

10. Sudoku Row Validator
Given 9×9 list:
● Check each row has numbers 1–9
● No duplicates
● Return valid/invalid

```python
def sudoku_row_validator(board):
    for row in board:
        # Check for duplicates
        if len(row) != len(set(row)):
            return "Invalid"

        #  Check numbers are from 1 to 9
        for num in row:
            if num < 1 or num > 9:
                return "Invalid"

    #  If all rows pass
    return "Valid"


# -------- USER INPUT --------
board = []

for i in range(9):
    row = list(map(int, input(f"Enter row {i+1} (9 numbers): ").split()))
    board.append(row)

# -------- FUNCTION CALL --------
result = sudoku_row_validator(board)
print(result)
```

16. Word Frequency + Longest Word

```python
sentence = input("Enter sentence: ")

words = sentence.split()
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Word count
print("Word Frequency:", freq)

# Longest word
longest = words[0]
for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)

# Most repeated word
most_repeated = words[0]
for word in freq:
    if freq[word] > freq[most_repeated]:
        most_repeated = word

print("Most repeated word:", most_repeated)
```

17. Nested List Sorting Without sort()
Sort 2D list by second value manually.

```python
data = [[1, 30], [2, 10], [3, 20]]

n = len(data)

for i in range(n):
    for j in range(0, n - i - 1):
        if data[j][1] > data[j + 1][1]:
            data[j], data[j + 1] = data[j + 1], data[j]

print(data)
```

```python
# @title
#problem based on ur age output the height in cms or inches.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Age data (NumPy array)
age = np.array([5, 8, 12, 15, 18, 21, 25]).reshape(-1, 1)

# Height data (NumPy array in cm)
height = np.array([110, 125, 140, 155, 165, 170, 175])

# Print dataset
print("Age:\n", age)
print("Height:\n", height)

# Train the model
model = LinearRegression()
model.fit(age, height)

# User input
user_age = int(input("Enter your age: "))
user_age = np.array([[user_age]])

# Predict height
predicted_height = model.predict(user_age)

# Output
print("Predicted Height:", predicted_height[0], "cm")



# Plotting the graph
plt.scatter(age, height, color="Blue") # Scatter original data
plt.plot(age, model.predict(age), color="Red") # Plot the regression line
plt.xlabel("Age")
plt.ylabel("Height (cm)")
plt.title("Age vs. Height Prediction")
plt.show()

#filter
height=np.array([110, 125, 140, 155])
updated_height=height[[True,False,True,False]]
print(updated_height)

```

```python

# 1. IMPORT LIBRARIES

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("LIBRARIES IMPORTED SUCCESSFULLY\n")


# 2. LOAD THE DATASETS
df1 = pd.read_csv('/content/Coffee1.csv')
df2 = pd.read_csv('/content/Coffee2.csv')

print("--1st OUTPUT : Coffee1 Dataset--")
print(df1)
print()

print("--2nd OUTPUT : Coffee2 Dataset--")
print(df2)
print()


# 3. NUMBER OF ROWS & COLUMNS

print("3. NUMBER OF ROWS AND COLUMNS")

print("Coffee1 Shape:", df1.shape)
print("Coffee2 Shape:", df2.shape)
print()


# 4. COLUMN NAMES
print("4. LIST OF COLUMNS")

print("Coffee1 Columns:")
print(df1.columns)
print()

print("Coffee2 Columns:")
print(df2.columns)
print()


# 5. DATA TYPES

print("5. DATA TYPES")

print("Coffee1 dtypes:")
print(df1.dtypes)
print()

print("Coffee2 dtypes:")
print(df2.dtypes)
print()


# 6. SUMMARY OF NUMERICAL COLUMNS

print("6. SUMMARY OF NUMERICAL COLUMNS")

print("Coffee1 Describe:")
print(df1.describe())
print()

print("Coffee2 Describe:")
print(df2.describe())
print()

# 7. CHECK FOR MISSING VALUES

print("7. MISSING / NULL VALUES")

print("Coffee1 Missing Values:")
print(df1.isnull().sum())
print()

print("Coffee2 Missing Values:")
print(df2.isnull().sum())
print()

# 8. FILL MISSING VALUES (NUMERIC → MEDIAN)

print("8. FILLING MISSING NUMERIC VALUES WITH MEDIAN")

df1.fillna(df1.median(numeric_only=True), inplace=True)
df2.fillna(df2.median(numeric_only=True), inplace=True)

print("Missing values after fillna (Coffee1):")
print(df1.isnull().sum())
print()

print("Missing values after fillna (Coffee2):")
print(df2.isnull().sum())
print()


# 9. FIRST & LAST 5 ROWS

print("9. FIRST 5 ROWS")
print(df1.head())
print()

print("LAST 5 ROWS")
print(df1.tail())
print()


# 10. GROUP BY ANALYSIS (EXAMPLE)
# (Change column names if needed)

print("10. GROUP BY ANALYSIS")

# Example: Group by a categorical column if present
cat_cols = df1.select_dtypes(include='object').columns

if len(cat_cols) > 0:
    col = cat_cols[0]
    print(f"Grouping Coffee1 by '{col}'")
    print(df1.groupby(col).count())
else:
    print("No categorical columns found for groupby")

print()

# 11. SAVE CLEANED DATASETS

df1.to_csv('Coffee1_Cleaned.csv', index=False)
df2.to_csv('Coffee2_Cleaned.csv', index=False)

print("11. CLEANED FILES SAVED SUCCESSFULLY\n")


# 12. VISUALIZATION – BAR CHART


if len(cat_cols) > 0:
    sns.countplot(x=cat_cols[0], data=df1)
    plt.title("Bar Chart of " + cat_cols[0])
    plt.xticks(rotation=45)
    plt.show()


# 13. HISTOGRAM

print("13. HISTOGRAM")

num_cols = df1.select_dtypes(include='number').columns

if len(num_cols) > 0:
    sns.histplot(df1[num_cols[0]], bins=30)
    plt.title("Histogram of " + num_cols[0])
    plt.show()


# 14. SCATTER PLOT

print("14. SCATTER PLOT")

if len(num_cols) >= 2:
    sns.scatterplot(x=num_cols[0], y=num_cols[1], data=df1)
    plt.title(f"{num_cols[0]} vs {num_cols[1]}")
    plt.show()

print("PROGRAM EXECUTED SUCCESSFULLY")
```

```python
print(" PROBLEM 1: COUNT OF RECORDS FOR EACH CATEGORY\n ")

# Step 1: Find all categorical (text) columns
categorical_columns = df1.select_dtypes(include='object').columns
print("Categorical columns found:", categorical_columns)
print()

# Step 2: Choose the first categorical column for analysis
selected_category = categorical_columns[0]
print("Selected category for analysis:", selected_category)
print()

# Step 3: Count how many times each category appears
category_count = df1[selected_category].value_counts()
print("Count of each category:")
print(category_count)
print()

# Step 4: Display the result using a bar chart
category_count.plot(kind='bar')
plt.title("Number of Records for Each " + selected_category)
plt.xlabel(selected_category)
plt.ylabel("Count")
plt.show()
```

```python
print("PROBLEM 2: AVERAGE VALUE FOR EACH CATEGORY\n")

# Step 1: Find categorical and numerical columns
categorical_columns = df1.select_dtypes(include='object').columns
numerical_columns = df1.select_dtypes(include='number').columns

print("Categorical columns:", categorical_columns)
print("Numerical columns:", numerical_columns)
print()

# Step 2: Select one categorical and one numerical column
category_column = categorical_columns[0]
number_column = numerical_columns[0]

print("Selected categorical column:", category_column)
print("Selected numerical column:", number_column)
print()

# Step 3: Calculate average value for each category
average_result = df1.groupby(category_column)[number_column].mean()
print("Average value for each category:")
print(average_result)
print()

# Step 4: Display the result using bar chart
average_result.plot(kind='bar')
plt.title("Average " + number_column + " for Each " + category_column)
plt.xlabel(category_column)
plt.ylabel("Average Value")
plt.show()
```

```python
print("PROBLEM 3: PERCENTAGE DISTRIBUTION USING PIE CHART\n")

# Step 1: Find categorical columns
categorical_columns = df1.select_dtypes(include='object').columns
print("Categorical columns found:", categorical_columns)
print()

# Step 2: Select one categorical column
selected_category = categorical_columns[0]
print("Selected category for pie chart:", selected_category)
print()

# Step 3: Count values in the selected category
category_values = df1[selected_category].value_counts()
print("Category counts:")
print(category_values)
print()

# Step 4: Show percentage distribution using pie chart
category_values.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Percentage Distribution of " + selected_category)
plt.ylabel("")   # Remove unnecessary label
plt.show()
```

```python
# multinomial regression.
# predict height based on the age.

import numpy as np
from sklearn.linear_model import LogisticRegression
import pandas as pd

#data
ages = np.array([2,5,7,10,12,14,16,18,20]).reshape(9,1)
height = np.array([74,84,75,110,115,105,130,150,145])
print(ages, height)

hclass=[]
for i in height:
  if i<100:
    hclass.append(0)#short
  elif i<=30:
      hclass.append(1)#tall
  else:
    hclass=np.array(hclass)

model = LogisticRegression()
model.fit(ages,height)


import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(ages,height,c = hclass, cmap = 'viridis',s=100)
plt.xlabel('age')
plt.ylabel('height')
plt.title("Realistic age vs height")
plt.colorbar()
plt.show()
# to create linespace
age_range = np.linspace(4,19,200).reshape(-1,1)
#creating Model.
pre = model.predict(age_range)
for i in range(len(pre)-1):
  if pre[i] != pre[i+1]:
    print('coundry near age ', age_range[i])

import matplotlib.pyplot as plt
plt.figure(figsize = (8,5))
plt.scatter(ages,height,c = hclass, cmap = 'viridis',s = 100)
plt.plot(age_range,pre,color = 'red')
plt.xlabel('age')
plt.ylabel('height')
plt.title('Realistic age vs height')
plt.colorbar()
plt.show()
```
