# Email Classification

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('email.csv')
```

```python
data
```

```python
data.shape
```

```python
# check missing values
data.isnull().sum()
```

```python
data.loc[:,'Category'].unique()
```

```python
data = data[data.loc[:,'Category']!='{"mode":"full"']
```

```python
# replace ham--> 1 ,spam-->0
data.loc[:,'class']=data['Category'].map({'ham':1,'spam':0})
```

```python
data.drop(columns=['Category'],inplace=True)
```

```python
import re
def clean_text(text):
  text = text.lower()
  text = re.sub(r'<.*?.',"",text)
  text =re.sub(r'http\S+|www\s+','',text)
  text = re.sub(r'[^x00-\x7F]+',' ',text)
  text=re.sub(r'[^a-z0-9\s]','',text)
  text = re.sub(r'\s+',' ',text).strip()
  return text
```

```python
data['cleaned_text']= data['Message'].apply(clean_text)
```

```python
# split data for x and y
x = data['cleaned_text']
y= data['class']
```

```python
# split data for training and tetsing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**CountVectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x_train_new = cv.fit_transform(x_train)
x_test_new = cv.transform(x_test)
```

```python
# fit_transform : learns and then transforms
# transform : no learning only transforming
```

#naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_new,y_train)
```

```python
y_pred = model.predict(x_test_new)
```

```python
# confusion matrix , accuracy , recall,precision,f1 score
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
msg = input("enter the message: ")
text = cv.transform([msg])
pred = model.predict(text)
if pred[0]==0:
  print("spam")
else:
  print("Ham")
```
